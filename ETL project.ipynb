```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import os
```

```python
url='https://www.ofx.com/en-us/forex-news/historical-exchange-rates/monthly-average-rates/'
response=requests.get(url)
soup=bs(response.text,'html.parser')
```

```python
# identify location of chromedriver and store it as a variable
driverPath = !which chromedriver
# Setup configuration variables to enable Splinter to interact with browser
executable_path = {'executable_path': driverPath[0]}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
res=soup.find_all('optgroup', label='All Currencies')[0]
option=res.find_all('option')
currencies=list(map(lambda o: o.text, option))
currencies
```

```python
currency_df=pd.DataFrame()
for currency in currencies:
    browser.visit(url)
    element=browser.find_by_xpath('//*[@id="#main"]/div[3]/div[1]/div/div/div/div[1]/div/div/div[1]').first
    element.click()
    currentcy_click=browser.find_by_text(currency)
    currentcy_click.click()
    button_click=browser.find_by_text('RETRIEVE DATA')
    button_click.click()
    temp_currency=pd.read_html(url)
    for index,row in temp_currency[0].iterrows():
#         currency_df['Date'].append(row[0])
#         currency_df['Rate'].append(row[1])
        currency_df = currency_df.append([[row[0],row[1],currency]])
    print(currency_df)
```

```python
currency_df
```

```python
currency_df=currency_df.rename(columns={0:'Date',1:'Rate',2:'Currency'})
currency_df
```

```python
currency_df=currency_df.reset_index()
currency_df=currency_df[['Date','Rate','Currency']]
currency_df
```
